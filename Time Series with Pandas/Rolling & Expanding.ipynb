# Rolling & Expanding

```python
# Allow for multiple outputs in same cell
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('starbucks.csv', index_col='Date', parse_dates=True)
df.head()
```

```python
df['Close'].plot(figsize=(12,5))
```

### Rolling

```python
df['Close'].plot(figsize=(12,5))
df.rolling(window=30).mean()['Close'].plot()
df.rolling(window=60).mean()['Close'].plot()
```

```python
df['Close: 30 Day Mean'] = df['Close'].rolling(window=30).mean()
```

```python
df[['Close', 'Close: 30 Day Mean']].plot(figsize=(12,5))
```

### Expanding

Instead of a moving window - what if you wanted to take into consideration all values up until the given point in time (back-end of the window isn't moving)

```python
df['Close'].expanding().mean().plot(figsize=(12,5))
```
