```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv(r'./macrodata.csv', index_col=0, parse_dates=True)
```

```python
df.head()
```

```python
df['realgdp'].plot(figsize=(12,5))
```

```python
from statsmodels.tsa.filters.hp_filter import hpfilter
```

```python
# lambda=1600 for quarterly data
# Lambda=6.25 for annual data
# Lambda=129,600 for monthly data
gdp_cycle, gdp_trend = hpfilter(df['realgdp'], lamb=1600)
```

```python
gdp_trend.plot()
gdp_cycle.plot()
```

```python
df['trend'] = gdp_trend
```

```python
df[['trend', 'realgdp']]['2005-01-01':].plot(figsize=(12,5))
```
