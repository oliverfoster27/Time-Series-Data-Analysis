```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv(r'./airline_passengers.csv', index_col=0, parse_dates=True)
df = df.dropna()
```

```python
df.head()
```

```python
df.plot()
```

### We have the option of either applying an addative or a multiplicative model. The trend looks to be exponential in the plot above so we'll apply multiplicitive

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
results = seasonal_decompose(df['Thousands of Passengers'], model='multiplicative')
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 12,5

results.plot();
```
